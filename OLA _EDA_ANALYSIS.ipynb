```python
# SQL Server connectivity
import pyodbc
from sqlalchemy import create_engine
```

```python
import pandas as pd
import pyodbc
from sqlalchemy import create_engine

engine = create_engine(
    "mssql+pyodbc://@localhost/OLA RIDE_ANALYTICS"
    "?driver=ODBC+Driver+17+for+SQL+Server"
    "&trusted_connection=yes"
)

pd.read_sql("SELECT DB_NAME() AS CurrentDB", engine)
```

```python
pd.read_sql("""
SELECT TABLE_NAME
FROM INFORMATION_SCHEMA.TABLES
WHERE TABLE_TYPE = 'BASE TABLE'
""", engine)
```

```python
df.to_sql(
    name="ola_rides",         
    con=engine,
    if_exists="replace",       
    index=False
)
```

```python
pd.read_sql("""
SELECT TABLE_NAME 
FROM INFORMATION_SCHEMA.TABLES
""", engine)
```

```python
pd.read_sql(
    "SELECT COUNT(*) AS total_rows FROM ola_rides",
    engine
)
```

```python
pd.read_sql(
    "SELECT TOP 5 * FROM ola_rides",
    engine
)
```
